```python
data = pd.read_csv("titanic_train.csv", dtype={
    "PassengerId": np.int16,
    "Survived": np.bool,
    "Pclass": np.int8,
    
    "Age": np.float16,
    "SibSp": np.int8,
    "Parch": np.int8,
    
     "Fare": np.float16,
    

})
data.info(memory_usage=True)
```

```python
data.Sex = data.Sex.astype("category")
```

```python
data.info(memory_usage=True)
```

```python
data.Sex
```

```python
data.Sex = data.Sex.cat.add_categories("unknown")
```

```python
data.info(memory_usage=True)
```

```python
data.Embarked = data.Embarked.astype("category")
```

```python
data.info(memory_usage=True)
```

```python
data.groupby("Sex").agg({"Age": np.mean})
```
